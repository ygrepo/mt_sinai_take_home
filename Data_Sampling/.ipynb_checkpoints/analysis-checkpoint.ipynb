# Calculate the Per Member Per Month Paid $ amount of Diabetic Patients vs. Non-Diabetic Patients

```python
%%sql select TOTAL/PATIENT_COUNT PMPM, case NMONTH
             when 1 then 'JAN'
             when 2 then 'FEB'
             when 3 then 'MAR'
             when 4 then 'APR'
             when 5 then 'MAY'
             when 6 then 'JUN'
             when 7 then 'JUL'
             when 8 then 'AUG'
             when 9 then 'SEP'
             when 10 then 'OCT'
             when 11 then 'NOV'
             when 12 then 'DEC'
             else 'UNK'
             end as MONTH
     from( select count(distinct claim.member_id) PATIENT_COUNT , sum(TCC_PAID) TOTAL,
        case str_split(DATE_OF_SERVICE, '-')[2] 
             when 'JAN' then 1
             when 'FEB' then 2
             when 'MAR' then 3
             when 'APR' then 4
             when 'MAY' then 5
             when 'JUN' then 6
             when 'JUL' then 7
             when 'AUG' then 8
             when 'SEP' then 9
             when 'OCT' then 10
             when 'NOV' then 11
             when 'DEC' then 12
             else -1
             end as NMONTH
        from CLAIM claim
        join MBR_ELIG elig on claim.MEMBER_ID = elig.MEMBER_ID
        where elig.DIABETES == 0
        group by NMONTH)
     order by NMONTH asc
```

```python
%%sql df << SELECT 
  CS_MBR_TMP.age, 
  CS_MBR_ELIG.DIABETES,
  SUM(CLAIM_HDR.TCC_PAID) AS TOTAL_COST
FROM 
  CLAIM_HDR 
  JOIN CS_MBR_TMP ON CLAIM_HDR.MEMBER_ID = CS_MBR_TMP.MEMBER_ID 
  JOIN CS_MBR_ELIG ON CLAIM_HDR.MEMBER_ID = CS_MBR_ELIG.MEMBER_ID
GROUP BY 
  CS_MBR_TMP.age, 
  CS_MBR_ELIG.DIABETES
ORDER BY 
  TOTAL_COST DESC, AGE;
```

```python
%%sql df << SELECT 
  CS_MBR_TMP.age, 
  CS_MBR_TMP.RACE_ETH, 
  CS_MBR_ELIG.DIABETES,
  SUM(CLAIM_HDR.TCC_PAID) AS TOTAL_COST
FROM 
  CLAIM_HDR 
  JOIN CS_MBR_TMP ON CLAIM_HDR.MEMBER_ID = CS_MBR_TMP.MEMBER_ID 
  JOIN CS_MBR_ELIG ON CLAIM_HDR.MEMBER_ID = CS_MBR_ELIG.MEMBER_ID
GROUP BY 
  CS_MBR_TMP.age, 
  CS_MBR_ELIG.DIABETES
ORDER BY 
  TOTAL_COST DESC, AGE;
```

```python
import altair as alt
import pandas as pd
import numpy as np

# create sample data
data = pd.DataFrame({
    'age': np.random.randint(18, 80, size=1000),
    'diabetes': np.random.choice(['Yes', 'No'], size=1000),
    'cost': np.random.normal(1000, 200, size=1000)
})

# create chart with mean cost by age and diabetes status
chart = alt.Chart(data).transform_aggregate(
    sum_cost='sum(cost)', groupby=['age', 'diabetes']
).mark_circle().encode(
    alt.X('age:Q', scale=alt.Scale(zero=False), axis=alt.Axis(title='Age')),
    alt.Y('sum_cost:Q', scale=alt.Scale(type='log'), axis=alt.Axis(title='Mean Cost ($)')),
    color=alt.Color('diabetes:N', scale=alt.Scale(scheme='category10'), legend=alt.Legend(title='Diabetes'))
)

# add trend line
trend = chart.transform_regression(
    'age', 'sum_cost', groupby=['diabetes'], method='poly', order=2
).mark_line().encode(
    alt.Color('diabetes:N', scale=alt.Scale(scheme='category10'))
)

# combine chart and trend line
chart = chart + trend

# display chart
chart
```

```python
import altair as alt
import pandas as pd
import numpy as np

# create sample data
data = pd.DataFrame({
    'age': np.random.randint(18, 80, size=1000),
    'gender': np.random.choice(['Male', 'Female'], size=1000),
    'race': np.random.choice(['White', 'Black', 'Asian', 'Hispanic'], size=1000),
    'disease': np.random.choice(['A', 'B', 'C'], size=1000),
    'cost': np.random.normal(1000, 200, size=1000)
})

# create scatter plot with color encoding for disease
scatter = alt.Chart(data).mark_circle().encode(
    alt.X('age:Q', scale=alt.Scale(zero=False), axis=alt.Axis(title='Age')),
    alt.Y('cost:Q', scale=alt.Scale(type='log'), axis=alt.Axis(title='Cost ($)')),
    color=alt.Color('disease:N', scale=alt.Scale(scheme='category10'), legend=alt.Legend(title='Disease'))
)

# create bar chart with color encoding for race
bars = alt.Chart(data).mark_bar().encode(
    alt.X('race:N', axis=alt.Axis(title='Race')),
    alt.Y('mean(cost):Q', axis=alt.Axis(title='Mean Cost ($)')),
    color=alt.Color('race:N', scale=alt.Scale(scheme='set1'), legend=None)
)

# create trend line with color encoding for gender and race
trend = alt.Chart(data).transform_regression(
    'age', 'cost', groupby=['gender', 'race'], method='poly', order=2
).mark_line().encode(
    alt.X('age:Q', axis=alt.Axis(title='Age')),
    alt.Y('cost:Q', axis=alt.Axis(title='Cost ($)')),
    color=alt.Color('gender:N', scale=alt.Scale(scheme='set1')),
    detail='race:N'
)

# combine charts using facet and layer
chart = alt.vconcat(
    scatter,
    alt.hconcat(bars, trend, spacing=20),
    #.facet(row='disease', data=data),
    spacing=20
)

# display chart
scatter
```

```python
import altair as alt
import pandas as pd
import numpy as np

# create sample data
data = pd.DataFrame({
    'x': np.random.randn(1000)
})

# create histogram layer with KDE
hist = alt.Chart(data).transform_density(
    'x', as_=['x', 'density']
).mark_area(opacity=0.5, interpolate='step').encode(
    alt.X('x:Q', bin=alt.Bin(maxbins=30), axis=alt.Axis(title='x')),
    alt.Y('density:Q', axis=alt.Axis(title='Density'))
)

# create trend line layer using the KDE
trend = hist.transform_regression(
    'x', 'density', method='linear', as_=['x', 'trend']
).mark_line(color='red').encode(
    alt.X('x:Q', axis=alt.Axis(title='x')),
    alt.Y('trend:Q', axis=alt.Axis(title='Trend'))
)

# combine layers
chart = hist + trend

# display chart
hist
```
